```python
#Q.2> Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

import random
import matplotlib.pyplot as plt

low = 1
high = 6
num_samples = 1000

samples = [random.randint(low, high) for _ in range(num_samples)]

outcomes = list(range(low, high + 1))  # All possible outcomes (1 to 6)
frequencies = [samples.count(outcome) for outcome in outcomes]

# Compute the probabilities (PMF)
probabilities = [freq / num_samples for freq in frequencies]

# Plot the PMF
plt.bar(outcomes, probabilities, color='skyblue', edgecolor='black')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of Discrete Uniform Distribution')
plt.xticks(outcomes)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Q.3> Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

def bernoulli_pdf(x, p):
    """
    Calculate the Probability Mass Function (PMF) for a Bernoulli distribution.

    Parameters:
    - x: The outcome (either 0 or 1)
    - p: The probability of success (1)

    Returns:
    - Probability for the given x (either 0 or 1)
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0  # If x is neither 0 nor 1, return 0 (invalid input)

# Example usage
p = 0.7  # Probability of success
x_values = [0, 1]  # Possible outcomes for Bernoulli distribution

for x in x_values:
    prob = bernoulli_pdf(x, p)
    print(f"P(X = {x}) = {prob}")
```

```python
#Q.6> Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution?

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1      # Lowest value
high = 6     # Highest value (inclusive)
num_samples = 10000  # Number of random samples

# Generate random samples from a discrete uniform distribution
samples = np.random.randint(low, high + 1, num_samples)

# Calculate the empirical CDF
outcomes = np.arange(low, high + 1)
cdf_values = [np.sum(samples <= outcome) / num_samples for outcome in outcomes]

# Plot the CDF
plt.step(outcomes, cdf_values, where='post', color='skyblue', linewidth=2, label='CDF')

# Add labels and title
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')

# Show grid
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
#Q.9> Write a Python function to calculate Z-scores from a dataset and plot them?

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    """
    Calculate the Z-scores for a given dataset.

    Parameters:
    - data: The dataset (list or numpy array).

    Returns:
    - z_scores: The Z-scores corresponding to the data.
    """
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev
    return z_scores

# Example usage
# Generate a random dataset (normal distribution for example)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate Z-scores for the dataset
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(8, 6))
plt.plot(z_scores, color='skyblue', label='Z-scores')
plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#Q.17> Generate random variables and calculate their corresponding probabilities using the binomial distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Function to generate random variables and calculate their corresponding probabilities using the binomial distribution
def binomial_distribution(n, p, size):
    # Generate random variables from the binomial distribution
    random_variables = np.random.binomial(n, p, size)

    # Calculate the probability mass function (PMF) for each random variable
    pmf_values = binom.pmf(random_variables, n, p)

    # Display the random variables and their corresponding probabilities
    for rv, prob in zip(random_variables, pmf_values):
        print(f"Random Variable: {rv}, Probability: {prob:.4f}")

    # Plot the PMF of the binomial distribution
    x = np.arange(0, n+1)  # Possible values of the random variable (from 0 to n)
    pmf_plot = binom.pmf(x, n, p)

    plt.figure(figsize=(8, 6))
    plt.stem(x, pmf_plot, basefmt=" ", use_line_collection=True)
    plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(x)
    plt.grid(True)
    plt.show()

# Example usage: Binomial distribution with n=10 trials, p=0.5 probability, and size=1000 random variables
n = 10  # Number of trials
p = 0.5  # Probability of success on each trial
size = 1000  # Number of random variables to generate

binomial_distribution(n, p, size)

```

```python
#Q.18> Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (data_point - mean) / std_dev
    return z_score, mean, std_dev

def plot_standard_normal(z_score):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, y, where=(x < z_score), color='orange', alpha=0.5, label=f'Z < {z_score:.2f}')
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')

    plt.title('Z-Score Position on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example data
data = np.random.normal(50, 10, 1000)  # Normal dataset: mean=50, std=10
data_point = 65  # Example point to evaluate

# Calculate Z-score
z, mean, std = calculate_z_score(data_point, data)
print(f"Data Point: {data_point}")
print(f"Mean: {mean:.2f}, Standard Deviation: {std:.2f}")
print(f"Z-score: {z:.2f}")

# Plot comparison
plot_standard_normal(z)
```

```python
#Q.19> Implement hypothesis testing using Z-statistics for a sample dataset?

import numpy as np
from scipy.stats import norm

def z_test(sample_mean, pop_mean, pop_std, n, alpha=0.05, alternative='less'):
    # Calculate Z statistic
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Calculate p-value based on the alternative hypothesis
    if alternative == 'less':
        p_value = norm.cdf(z)
        critical_z = norm.ppf(alpha)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z)
        critical_z = norm.ppf(1 - alpha)
    else:  # two-tailed
        p_value = 2 * min(norm.cdf(z), 1 - norm.cdf(z))
        critical_z = norm.ppf(1 - alpha / 2)

    # Decision
    reject_null = p_value < alpha

    # Output results
    print(f"Z-statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-value: {critical_z:.4f}")
    print("Conclusion:", "Reject H0" if reject_null else "Fail to reject H0")

# Example values
sample_mean = 980
pop_mean = 1000
pop_std = 80
n = 50
alpha = 0.05

# Run the Z-test
z_test(sample_mean, pop_mean, pop_std, n, alpha, alternative='less')
```

```python
#Q.20> Create a confidence interval for a dataset using Python and interpret the result?

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error = s / sqrt(n)
    df = n - 1  # Degrees of freedom

    # Get t critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df)

    # Margin of error
    margin = t_crit * std_err

    # Confidence interval
    lower_bound = mean - margin
    upper_bound = mean + margin

    # Output
    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print("Interpretation: We are", int(confidence * 100), "% confident that the true population mean lies within this interval.")

# Generate sample data (e.g., exam scores)
np.random.seed(42)
data = np.random.normal(loc=75, scale=10, size=30)

# Calculate and print confidence interval
confidence_interval(data, confidence=0.95)
```

```python
#Q.21> Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?

import numpy as np
from scipy import stats

def generate_data_and_ci(mean=100, std_dev=15, n=50, confidence=0.95):
    # Step 1: Generate data from a normal distribution
    np.random.seed(0)  # For reproducibility
    data = np.random.normal(loc=mean, scale=std_dev, size=n)

    # Step 2: Calculate sample statistics
    sample_mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    df = n - 1  # degrees of freedom for t-distribution

    # Step 3: Get t critical value for the confidence level
    t_crit = stats.t.ppf((1 + confidence) / 2, df)

    # Step 4: Calculate margin of error and confidence interval
    margin = t_crit * std_err
    lower_bound = sample_mean - margin
    upper_bound = sample_mean + margin

    # Step 5: Output the results
    print(f"Generated Data (first 5): {data[:5]}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print(f"Interpretation: We are {int(confidence * 100)}% confident that the true mean lies within this interval.")

# Run the function
generate_data_and_ci()
```

```python
#Q.22 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std_dev=1):
    # Generate x values around the mean ± 4 standard deviations
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)

    # Calculate the PDF values
    pdf = norm.pdf(x, loc=mean, scale=std_dev)

    # Plot the PDF
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, color='blue', lw=2, label=f'N({mean}, {std_dev}²)')
    plt.title('Normal Distribution - Probability Density Function (PDF)')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.axvline(mean, color='red', linestyle='--', label='Mean')
    plt.fill_between(x, pdf, alpha=0.2)  # Fill under the curve
    plt.legend()
    plt.show()

# Example usage: standard normal distribution
plot_normal_pdf(mean=0, std_dev=1)
```

```python
#Q.23>  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam=5, max_x=15):
    # Generate range of x values
    x = np.arange(0, max_x + 1)

    # Calculate CDF values
    cdf_values = poisson.cdf(x, mu=lam)

    # Print and interpret CDF for a specific x
    x_check = 7
    prob = poisson.cdf(x_check, mu=lam)
    print(f"P(X ≤ {x_check}) = {prob:.4f}")
    print(f"Interpretation: There is a {prob*100:.2f}% chance that the number of events is ≤ {x_check} when λ = {lam}.")

    # Plot the CDF
    plt.figure(figsize=(8, 5))
    plt.step(x, cdf_values, where='post', label=f'Poisson CDF (λ={lam})', color='purple')
    plt.title('Poisson Distribution - Cumulative Distribution Function (CDF)')
    plt.xlabel('x (Number of Events)')
    plt.ylabel('P(X ≤ x)')
    plt.grid(True)
    plt.xticks(x)
    plt.ylim(0, 1.05)
    plt.legend()
    plt.show()

# Example usage
plot_poisson_cdf(lam=5, max_x=15)

```

```python
#Q.24> Simulate a random variable using a continuous uniform distribution and calculate its expected value?

import numpy as np

def simulate_uniform_continuous(a=10, b=20, n=10000):
    # Simulate n random variables from a Uniform(a, b) distribution
    data = np.random.uniform(low=a, high=b, size=n)

    # Calculate theoretical expected value
    expected_value = (a + b) / 2

    # Calculate empirical (simulated) mean
    empirical_mean = np.mean(data)

    # Display results
    print(f"Uniform Distribution U({a}, {b})")
    print(f"Theoretical Expected Value: {expected_value:.2f}")
    print(f"Empirical Mean (from simulation): {empirical_mean:.2f}")

# Run the simulation
simulate_uniform_continuous(a=10, b=20, n=10000)

```

```python
#Q.25> Write a Python program to compare the standard deviations of two datasets and visualize the difference?

import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations():
    # Generate two datasets
    np.random.seed(42)
    data1 = np.random.normal(loc=50, scale=5, size=1000)   # Mean=50, SD=5
    data2 = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, SD=15

    # Calculate standard deviations
    std1 = np.std(data1)
    std2 = np.std(data2)

    # Print comparison
    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")

    # Plot histograms
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.hist(data1, bins=30, alpha=0.6, label=f'Dataset 1 (SD={std1:.2f})', color='blue')
    plt.hist(data2, bins=30, alpha=0.6, label=f'Dataset 2 (SD={std2:.2f})', color='orange')
    plt.title('Histograms of Both Datasets')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

    # Plot standard deviations
    plt.subplot(1, 2, 2)
    plt.bar(['Dataset 1', 'Dataset 2'], [std1, std2], color=['blue', 'orange'])
    plt.title('Comparison of Standard Deviations')
    plt.ylabel('Standard Deviation')

    plt.tight_layout()
    plt.show()

# Run the comparison
compare_standard_deviations()


```

```python
#Q.26> Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?

import numpy as np

def calculate_range_and_iqr(mean=50, std_dev=10, n=1000):
    # Generate dataset from normal distribution
    np.random.seed(0)
    data = np.random.normal(loc=mean, scale=std_dev, size=n)

    # Calculate Range
    data_range = np.max(data) - np.min(data)

    # Calculate IQR
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    # Output the results
    print(f"Dataset Summary (Normal Distribution N({mean}, {std_dev}²))")
    print(f"Range: {data_range:.2f}")
    print(f"Q1 (25th percentile): {q1:.2f}")
    print(f"Q3 (75th percentile): {q3:.2f}")
    print(f"Interquartile Range (IQR): {iqr:.2f}")

# Run the function
calculate_range_and_iqr()

```

```python
#Q.27> Implement Z-score normalization on a dataset and visualize its transformation?

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization():
    # Step 1: Generate original data (e.g., from normal distribution)
    np.random.seed(42)
    data = np.random.normal(loc=100, scale=20, size=1000)

    # Step 2: Apply Z-score normalization
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev

    # Step 3: Print summary
    print(f"Original Data: Mean = {mean:.2f}, Std Dev = {std_dev:.2f}")
    print(f"Normalized Data: Mean = {np.mean(z_scores):.2f}, Std Dev = {np.std(z_scores):.2f}")

    # Step 4: Plot before and after normalization
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, color='skyblue', edgecolor='black')
    plt.title('Original Data Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
    plt.title('Z-score Normalized Data')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Run it
z_score_normalization()
```

```python
#Q.28> Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution?

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std_dev=1, size=1000, excess=True):
    # Generate normally distributed data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness
    data_skewness = skew(data)

    # Calculate kurtosis
    data_kurtosis = kurtosis(data, fisher=excess)  # fisher=True gives excess kurtosis

    # Print results
    print("Generated Normal Distribution Data")
    print(f"Skewness       : {data_skewness:.4f} {'(≈ 0)' if abs(data_skewness) < 0.1 else '(Not symmetric)'}")
    print(f"{'Excess ' if excess else ''}Kurtosis     : {data_kurtosis:.4f} {'(≈ 0)' if excess and abs(data_kurtosis) < 0.1 else ''}")

# Run the function
analyze_normal_distribution_skew_kurtosis()
```
